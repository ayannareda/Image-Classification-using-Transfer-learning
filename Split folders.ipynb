```python
import os
import numpy as np
import shutil
import random
```

```python
root_dir = 'C:/Users/Mechanic/Transfer Learning/Dataset/'
classes_dir = ['owl',
            'galaxy',
            'lightning',
            'wine-bottle',
            't-shirt',
            'waterfall',
            'sword',
            'school-bus',           
            'calculator',
            'sheet-music',            
            'airplanes',     
            'lightbulb',
            'skyscraper',   
            'mountain-bike',         
            'fireworks',       
            'computer-monitor',              
            'bear',
            'grand-piano',
            'kangaroo',
            'laptop',]

val_ratio = 0.25
test_ratio = 0.25

rot_dir = 'C:/Users/Mechanic/Transfer Learning/Data/'

for cls in classes_dir:
    os.makedirs(rot_dir +'train/' + cls)
    os.makedirs(rot_dir +'val/' + cls)
    os.makedirs(rot_dir +'test/' + cls)

    src = root_dir + cls 
    
    allFileNames = os.listdir(src)
    train_FileNames, val_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                          [int(len(allFileNames)* (1 - (val_ratio + test_ratio))), 
                                                           int(len(allFileNames)* (1 - test_ratio))])

    train_FileNames = [src+'/'+ name for name in train_FileNames.tolist()]
    val_FileNames = [src+'/' + name for name in val_FileNames.tolist()]
    test_FileNames = [src+'/' + name for name in test_FileNames.tolist()]

    print('Total images: ', len(allFileNames))
    print('Training: ', len(train_FileNames))
    print('Validation: ', len(val_FileNames))
    print('Testing: ', len(test_FileNames))

    for name in train_FileNames:
        shutil.copy(name, rot_dir +'train/' + cls)

    for name in val_FileNames:
        shutil.copy(name, rot_dir +'val/' + cls)
        
    for name in test_FileNames:
        shutil.copy(name, rot_dir +'test/' + cls)
```
